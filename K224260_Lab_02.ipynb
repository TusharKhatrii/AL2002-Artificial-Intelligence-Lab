```python
import random

class Environment:
    def __init__(self):
        # Initialize system components A to I with random statuses
        self.components = {chr(65 + i): random.choice(['Safe', 'Vulnerable']) for i in range(9)}

    def display_system_state(self, message):
        print(f"\n{message}")
        for component, status in self.components.items():
            print(f"{component}: {status}")

class SecurityAgent:
    def __init__(self, environment):
        self.environment = environment
        self.vulnerable_components = []

    def scan_system(self):
        print("\nScanning system...")
        for component, status in self.environment.components.items():
            if status == "Vulnerable":
                print(f"Warning: {component} is Vulnerable.")
                self.vulnerable_components.append(component)
            else:
                print(f"Success: {component} is Safe.")

    def patch_vulnerabilities(self):
        print("\nPatching vulnerabilities...")
        for component in self.vulnerable_components:
            print(f"Patching {component}...")
            self.environment.components[component] = "Safe"
        self.vulnerable_components.clear()

# Simulation Execution
environment = Environment()
security_agent = SecurityAgent(environment)

# Step 1: Initial System Check
environment.display_system_state("Initial System State:")

# Step 2: System Scan
security_agent.scan_system()

# Step 3: Patching Vulnerabilities
security_agent.patch_vulnerabilities()

# Step 4: Final System Check
environment.display_system_state("Final System State:")
```

```python
import random
class server:
  def __init__(self):
    self.servers = {i+1: random.choice(['Underload','Balanced','Overloaded']) for i in range(5)}

  def display(self,message):
      print(f"{message}")
      for key,val in self.servers.items():
        print(f"Server {key}: {val}")

class LoadBalancer:
  def __init__(self):
    self.underloaded = []
    self.overloaded = []
    pass

  def scan(self,server):
    for key,val in server.servers.items():
      if val == 'Underload':
        self.underloaded.append(key)
      elif val == 'Overloaded':
        self.overloaded.append(key)

    if not self.overloaded:
      print("No balancing possible!")
      return

    if  not self.underloaded:
      print("No balancing required!")
      return

    for i in range(min(len(self.overloaded), len(self.underloaded))):
      print(f"Balancing server {self.overloaded[i]} and server {self.underloaded[i]}")
      server.servers[self.overloaded[i]] = 'Balanced'
      server.servers[self.underloaded[i]] = 'Balanced'


def runAgent(server,loadbalancer):
  loadbalancer.scan(server)




server = server()
loadbalancer = LoadBalancer()
server.display("Initial state")
runAgent(server,loadbalancer)
server.display("Final state")
```

```python
import random
class server:
  def __init__(self,num):
    self.backup = [random.choice(['Completed','Failed']) for i in range(num)]

  def display(self,message):
      print(f"{message}:")
      for i in range(len(self.backup)):
        print(f"{i+1}: {self.backup[i]}")

class BackupAgent:
  def __init__(self):
    pass

  def scan(self,server):
    for i in range(len(server.backup)):
      if server.backup[i] == 'Failed':
        print(f"Retrying backup {i}....")
        print(f"Backup completed")
        server.backup[i] = 'Completed'


def runAgent(server,BackupAgent):
  BackupAgent.scan(server)


server = server(5)
backupAgent = BackupAgent()
server.display("Initial state")
runAgent(server,backupAgent)
server.display("Final state")
```

```python
import random

class Environment:
    def __init__(self):
        self.components = {chr(65 + i): random.choice(['Safe', 'Low Risk Vulnerable', 'High Risk Vulnerable']) for i in range(9)}

    def display(self, message):
        print(f"\n{message}:")
        for component, status in self.components.items():
            print(f"Component {component}: {status}")


class SecurityAgent:
    def __init__(self):
        self.low_risk_vulnerable = []
        self.high_risk_vulnerable = []

    def scan_and_patch(self, environment):
        print("\nScanning system components...")

        # System Scan
        for component, status in environment.components.items():
            if status == 'Safe':
                print(f"{component} is Safe. No action needed.")
            elif status == 'Low Risk Vulnerable':
                print(f"WARNING: {component} has a Low Risk Vulnerability.")
                self.low_risk_vulnerable.append(component)
            elif status == 'High Risk Vulnerable':
                print(f"ALERT: {component} has a High Risk Vulnerability. Premium service required.")
                self.high_risk_vulnerable.append(component)

        # patch
        for component in self.low_risk_vulnerable:
            print(f"Patching Low Risk Vulnerability in {component}...")
            environment.components[component] = 'Safe'

        print("\nPatching complete.")


def runAgent(env,agent):
  agent.scan_and_patch(env)


environment = Environment()
agent = SecurityAgent()
environment.display("Initial System State")
runAgent(environment,agent)
environment.display("Final System State")

```

```python
import random

class Environment:
    def __init__(self):
        self.rooms = {f"Room {i}": None for i in range(1, 6)}
        self.medicine_storage = ["Painkillers", "Antibiotics", "Insulin", "Sedatives"]
        self.nurse_station = "Nurse Station"

    def update_room(self, room_number, medicine):
        self.rooms[room_number] = medicine

    def display(self, message):
        print(f"\n{message}:")
        for room, medicine in self.rooms.items():
            print(f"{room}: {medicine if medicine else 'Empty'}")
        print(f"Medicine Storage: {', '.join(self.medicine_storage)}")


class DeliveryRobot:
    def __init__(self):
        self.goal = None

    def set_goal(self, room_number, medicine):
        self.goal = (room_number, medicine)

    def move_to(self, location):
        print(f"Moving to {location}...")

    def pick_up_medicine(self):
        print(f"Picking up {self.goal[1]} from storage...")

    def scan_patient_id(self, room_number):
        print(f"Scanning Patient ID in {room_number}... Verification successful!")

    def deliver_medicine(self):
        print(f"Delivering {self.goal[1]} to {self.goal[0]}...")

    def alert_staff(self):
        print("Alerting staff at Nurse Station...")

    def execute_goal(self, environment):
        if self.goal:
            self.move_to("Medicine Storage")
            self.pick_up_medicine()
            self.move_to(self.goal[0])
            self.scan_patient_id(self.goal[0])
            self.deliver_medicine()
            environment.update_room(self.goal[0], self.goal[1])
            print(f"Delivery to {self.goal[0]} complete.")
            self.goal = None


def runAgent(environment, robot):
    robot.set_goal("Room 3", "Painkillers")
    robot.alert_staff()
    robot.execute_goal(environment)


environment = Environment()
robot = DeliveryRobot()
environment.display("Initial Hospital Layout")
runAgent(environment, robot)
environment.display("Final Hospital Layout")
```

```python
import random

class Environment:
    def __init__(self):
        self.rooms = {chr(97+i): random.choice([' ', 'ðŸ”¥']) for i in range(9)}  # range(10) to include 'j'

    def display(self):

        print("\nCurrent Environment Status:")
        for key, val in self.rooms.items():
            print(f"{key}: {val}")

    def extinguish_fire(self, room):
        # If there's a fire in the room, extinguish it and update the room's status
        if self.rooms[room] == 'ðŸ”¥':
            self.rooms[room] = ' '  # Fire is extinguished, room becomes safe
            print(f"Fire extinguished in room {room}!")


class FirefightingRobot:
    def __init__(self, environment):
        self.environment = environment
        self.path = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
        self.current_room = 'a'

    def move(self):
        # Move the robot along the predefined path
        for room in self.path:
            print(f"\nMoving to room {room}...")
            self.current_room = room
            self.environment.display()
            self.check_and_extinguish_fire()

    def check_and_extinguish_fire(self):
        # Check if the current room has a fire and extinguish it
        if self.environment.rooms[self.current_room] == 'ðŸ”¥':
            self.environment.extinguish_fire(self.current_room)
        else:
            print(f"Room {self.current_room} is safe, moving to the next room...")


def runAgent(robot):
  robot.move()


environment = Environment()
robot = FirefightingRobot(environment)
environment.display()
runAgent(robot)
print("All fires extingueshed")
```
